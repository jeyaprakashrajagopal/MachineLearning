```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from pathlib import Path

# Project root 3 levels up
project_root = Path.cwd().parents[2]
```

```python
df = pd.read_csv(project_root / 'datasets/01_regression/Position_Salaries.csv')
df.head()
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
X.shape, y.shape
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
X_poly
```

```python
linear_regression_1 = LinearRegression()
linear_regression_1.fit(X_poly, y)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X, y)
plt.plot(X, linear_regression.predict(X))
plt.show
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X, y)
plt.plot(X, linear_regression_1.predict(poly_reg.fit_transform(X)))
plt.show()
```
