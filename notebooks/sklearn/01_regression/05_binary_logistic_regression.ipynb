```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

### Step-1: Data preprocessing

```python
X,y = make_blobs(n_samples=2000, n_features=2, centers=2, cluster_std=3, random_state=42)
```

```python
def visualize(X,y):
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.show()
```

```python
visualize(X,y)
```

```python
standard_scalar = StandardScaler()
```

```python
standard_scalar.fit(X=X)
X = standard_scalar.transform(X=X)
```

```python
visualize(X, y)
```

```python
XT,Xt,yT,yt=train_test_split(X,y,test_size=0.25,shuffle=False,random_state=0)
```

### Step-2: Model training

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(XT, yT)
```

### Step-3: Prediction

```python
y_predict = logistic_regression.predict(XT)
```

### Step-4: Accuracy
